```
#Importing the libraries
from pyspark.sql import SparkSession
```

```
#Creating a new spark session
spark = SparkSession.builder.appName('ops').getOrCreate()
```

```
#Reading the csv file from DBFS with options of headers and inferschema
df1 = spark.read.format("csv").option("header",True).option("inferSchema",True).load("dbfs:/FileStore/shared_uploads/yashbhatia96@g.ucla.edu/appl_stock.csv")
```

```
#Printing the df, shows only first 20 rows
df1.show()
```

```
#Using "filter" to filter out rows based on the condition (Close<200) and using ".select" to select only few columns
df1.filter("Close < 200").select(['Date','Open','Close','Volume']).show()
```

```
#Multiple conditions (USE & operator to put an "and" condition and put both conditions in brackets otherwise you will get an error)
df1.filter((df1['Close'] < 200) & (df1['Open'] > 200)).select(['Date','Open','Close','Volume']).show()
```

```
#Collect operator to fetch result and store it
#Collect operator will return a list of row objedts
#result = df1.filter("Low == 197.16").collect() -- sql way
result = df1.filter(df1['Low'] == 197.16).collect()
type(result)
```

```
#Storing first row object in "row" variable
row = result[0]
```

```
#Use asDict() method to present the row object as a Dictionary
row.asDict()
```

```
#We can select the column we need from the Dictionary variable
row.asDict()['Volume']
```
