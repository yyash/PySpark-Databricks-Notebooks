```
#Importing SparkSession class from pyspark.sql library
#Building the spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Basics").getOrCreate()
```

```
#Reading the file from DBFS with header and inferSchema option set as True
df = spark.read.format("csv").option('header',True).option('inferSchema', True).load("dbfs:/FileStore/shared_uploads/yashbhatia96@g.ucla.edu/appl_stock.csv")
```

```
#Importing date time functions
from pyspark.sql.functions import (dayofmonth, hour, dayofyear, month, year, weekofyear, format_number, date_format)
```

```
df.select(dayofmonth('Date')).show()
df.select(hour('Date')).show()
```

```
#Avg close per year
df_withYear = df.withColumn('Year',year('Date'))
result = df_withYear.groupBy('Year').mean().select(['Year','avg(Close)'])
```

```
result = result.withColumnRenamed('avg(Close)', 'Average Closing per Year')
result.select('Year', format_number('Average Closing per Year',2).alias('Average Closing per Year')).show()
```
