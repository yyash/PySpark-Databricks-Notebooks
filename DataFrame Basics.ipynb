```
from pyspark.sql.types import (StructField,StringType,IntegerType,StructType)
```

```
data_schema = [StructField('age',IntegerType(),True),
              StructField('name',StringType(),True)]
```

```
final_struct = StructType(fields=data_schema)
```

```
df1_new_schema = spark.read.schema(final_struct).json("dbfs:/FileStore/shared_uploads/yashbhatia96@g.ucla.edu/people-1.json")
```

```
df1_new_schema.describe()
```

```
df1_new_schema.printSchema()
```

```
df1_new_schema.show()
```

```
df1_new_schema.select('age').show()
```

```
type(df1_new_schema.head(2)[0])
```

```
df1_new_schema.select(['age','name']).show()
```

```
#Creating a new column (Copying the column AGE)
#The below command does not create a new column in the dataframe.
#It just shows how the column would look.
#You need to assign the below statement to a variable for it to be saved
df1_new_schema.withColumn('doubleAge',df1_new_schema['age']*2).show()
```

```
df1_new_schema.show()
```

```
#Renaming an existing column (Copying the column AGE)
#The below command does not rename the column in the dataframe.
#It just shows how the column would look.
#You need to assign the below statement to a variable for it to be saved
df1_new_schema.withColumnRenamed('age','new_age').show()
```

```
df1_new_schema.show()
```

```
#Using Sql on Dataframes by creating a temporary table view of the df
df1_new_schema.createOrReplaceTempView('people')
```

```
#Use spark.sql to query the df as a table and show the result
results = spark.sql("select * from people")
results.show()
```

```
#Another example of a query
new_results = spark.sql("select * from people where age=30")
new_results.show()
```
