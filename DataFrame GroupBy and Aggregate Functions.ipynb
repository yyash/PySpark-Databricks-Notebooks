```
#Importing SparkSession class from pyspark.sql library
from pyspark.sql import SparkSession
```

```
#Building the spark session
spark = SparkSession.builder.appName('groupbyagg').getOrCreate()
```

```
#Reading the file from DBFS with header and inferSchema option set as True
df = spark.read.format("csv").option('header',True).option('inferSchema', True).load("dbfs:/FileStore/shared_uploads/yashbhatia96@g.ucla.edu/sales_info.csv")
```

```
df.show()
```

```
#Print schema for df
df.printSchema()
```

```
#Using groupBy command to show various aggregation functions
df.groupBy('Company').mean().show()
df.groupBy('Company').sum().show()
df.groupBy('Company').max().show()
df.groupBy('Company').min().show()
df.groupBy('Company').count().show()
```

```
#Performing aggregation over the entire dataset
df.agg({'Sales':'sum'}).show()
df.agg({'Sales':'max'}).show()
```

```
group_data = df.groupBy('Company')
group_data.agg({'Sales':'max'}).show()
```

```
#Importing functions in pyspark
from pyspark.sql.functions import countDistinct,avg,stddev
df.select(countDistinct('Sales')).show()
df.select(avg('Sales').alias('Average_Sales')).show()
```

```
#Formatting the result using format_number function
from pyspark.sql.functions import format_number
standard_dev = df.select(stddev('Sales').alias('std'))
standard_dev.select(format_number('std',2).alias('std')).show()
```

```
#Order data in ascending order
df.orderBy('Sales').show()
#Order data in descending order
df.orderBy(df['Sales'].desc()).show()
```
